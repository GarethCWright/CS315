```python
"""
Output: the time taken for the function to run
"""
def function_timer(func, *args, **kwargs):
      start = time.time()
      func(*args, **kwargs)
      end = time.time()
      return float(end)-float(start)
```

```python
import random
def quickselect(A, i):
  if len(A) == 1:
    return A[0]
  pivot = random.choice(A)
  lows = [x for x in A if x < pivot]
  highs = [x for x in A if x > pivot]
  pivots = [x for x in A if x == pivot]
  if i <= len(lows):
    return quickselect(lows, i)
  elif i <= len(lows) + len(pivots):
    return pivot
  else:
    return quickselect(highs, i - len(lows) - len(pivots))
```

```python
#QUICK-SELECT WITH RANDOM ARRAYS

MAX_TRIALS = 7  #number of sizes of arrays to test; 10^1 -> 10^(MAX_TRIALS-1)
times = []

for i in range(1, MAX_TRIALS):
    size = 10**i
    avg_time = 0

    #avg it out over 4 trials
    for j in range(4):
        rand_arr = list(range(size))
        random.shuffle(rand_arr)
        avg_time += function_timer(quickselect, rand_arr, size//2)

    #compute the average time for that size of array
    times.append(avg_time / 4)

#print times
for i in range(MAX_TRIALS-1):
    print(f"Array size 10^{i+1}: {times[i]} seconds")

#RATIO TEST
print("Ratio test w/ Random Arrays (time / n):")
for i in range(MAX_TRIALS-1):
    ratio = float(times[i]) / float(10**(i+1))
    print(f"For array length 10^{i+1}: {ratio}")
```

```python
#QUICK-SELECT WITH SEQUENTIAL ARRAYS

MAX_TRIALS = 8  #number of sizes of arrays to test; 10^1 -> 10^(MAX_TRIALS-1)
times = []

for i in range(1, MAX_TRIALS):
    size = 10**i
    avg_time = 0

    #avg it out over 4 trials
    for j in range(4):
        seq_arr = list(range(size))  # changed from random to sequential
        avg_time += function_timer(quickselect, seq_arr, size//2)

    #compute the average time for that size of array
    times.append(avg_time / 4)

#print times
for i in range(MAX_TRIALS-1):
    print(f"Array size 10^{i+1}: {times[i]} seconds")

#RATIO TEST
print("Ratio test w/ Sequential Arrays (time / n^2):")
for i in range(MAX_TRIALS-1):
    ratio = float(times[i]) / (float(10**(i+1))**2)
    print(f"For array length 10^{i+1}: {ratio}")

```

```python
# QUICK-SELECT WITH BACKWARDS SEQUENTIAL ARRAYS

MAX_TRIALS = 7  #number of sizes of arrays to test; 10^1 -> 10^(MAX_TRIALS-1)
times = []

for i in range(1, MAX_TRIALS):
    size = 10**i
    avg_time = 0

    #avg it out over 4 trials
    for j in range(4):
        rev_seq_arr = list(range(size-1, -1, -1))  # changed from random to reverse sequential
        avg_time += function_timer(quickselect, rev_seq_arr, size//2)

    #compute the average time for that size of array
    times.append(avg_time / 4)

#print times
for i in range(MAX_TRIALS-1):
    print(f"Array size 10^{i+1}: {times[i]} seconds")

#RATIO TEST
print("Ratio test w/ Random Arrays (time / n^2):")
for i in range(MAX_TRIALS-1):
    ratio = float(times[i]) / (float(10**(i+1))**2)
    print(f"For array length 10^{i+1}: {ratio}")
```
